```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('pred_obs_field_long.csv', sep = ';')
df.head()
```

```python
# Plot graph by field
p1 = sns.catplot(data=df, kind="bar", y="Field_Name",x='Ctotal', hue="Type", palette=['pink', 'brown'], height=6,
    orient = 'h')
p1.set_axis_labels("Total Carbon (%)", "Fields")
p1.legend.set_title("")
plt.savefig('CTotalFields_allyears_average.jpeg')
plt.show()
```

```python
# Plot graph by farmlet
p2 = sns.catplot(data=df, kind="bar", y="Farmlet",x='Ctotal', hue="Type", palette=['pink', 'brown'], height=6,
    orient = 'h')
p2.set_axis_labels("Total Carbon (%)", 'Farmlets')
p2.legend.set_title("")
plt.savefig('CTotalFarmlet_allyears_average.jpeg')
plt.show()
```

```python
# Plot graph by farmlet and by year
p3 = sns.FacetGrid(data=df, col="Year", aspect=.7, col_wrap = 3) 
p3.map(sns.barplot, "Farmlet", "Ctotal", "Type", palette=['pink', 'brown'])
p3.set_axis_labels("Farmlet", "Total Carbon (%)")
#plt.savefig('CTotalFarmlet_allyears.jpeg')
plt.show()
```
