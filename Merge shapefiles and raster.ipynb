### Merge shapefiles with csv

```python
# Import libraries
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load area and rename column
area = gpd.read_file('C:/Users/neliq/Documents/NIF/Rothamsted/shapefiles/area_fieldCTotal.shp')
area.rename(columns = {'Field_Name': 'Field'}, inplace = True)
area.head()
```

```python
# Load csv
year_2012 = pd.read_csv('C:/Users/neliq/Documents/NIF/Rothamsted/datasets_RR/data_with_RS/data_RS_by_month/definitive/1.filesL8_2012_mean.csv',
                       sep= ';', usecols = ['Field', 'C_total_%w/w'])
year_2012.rename(columns = {'C_total_%w/w': 'CTotal2012'}, inplace = True)

year_2016 = pd.read_csv('C:/Users/neliq/Documents/NIF/Rothamsted/datasets_RR/data_with_RS/data_RS_by_month/definitive/2.filesL8_2016_mean.csv',
                       sep= ';', usecols = ['Field', 'C_total_%w/w'])
year_2016.rename(columns = {'C_total_%w/w': 'CTotal2016'}, inplace = True)

year_2018 = pd.read_csv('C:/Users/neliq/Documents/NIF/Rothamsted/datasets_RR/data_with_RS/data_RS_by_month/definitive/4.filesL8_2018_mean.csv',
                       sep= ';', usecols = ['Field', 'C_total_%w/w'])
year_2018.rename(columns = {'C_total_%w/w': 'CTotal2018'}, inplace = True)

year_2019 = pd.read_csv('C:/Users/neliq/Documents/NIF/Rothamsted/datasets_RR/data_with_RS/data_RS_by_month/definitive/6.filesL8_2019_mean.csv',
                       sep= ';', usecols = ['Field', 'C_total_%w/w'])
year_2019.rename(columns = {'C_total_%w/w': 'CTotal2019'}, inplace = True)

year_2020 = pd.read_csv('C:/Users/neliq/Documents/NIF/Rothamsted/datasets_RR/data_with_RS/data_RS_by_month/definitive/8.filesL8_2020_mean.csv',
                       sep= ';', usecols = ['Field', 'C_total_%w/w'])
year_2020.rename(columns = {'C_total_%w/w': 'CTotal2020'}, inplace = True)

year_2021 = pd.read_csv('C:/Users/neliq/Documents/NIF/Rothamsted/datasets_RR/data_with_RS/data_RS_by_month/definitive/10.filesL8_2021_mean.csv',
                       sep= ';', usecols = ['Field', 'C_total_%w/w'])
year_2021.rename(columns = {'C_total_%w/w': 'CTotal2021'}, inplace = True)
```

```python
# Join all the csv
new_df = year_2012.join([year_2016['CTotal2016'], year_2018['CTotal2018'], year_2019['CTotal2019'], year_2020['CTotal2020'],
                        year_2021['CTotal2021']])
new_df.head()
```

```python
# Merge by field and export
area_withCT = gpd.GeoDataFrame(area.merge(new_df, on = 'Field'))
area_withCT.to_file('C:/Users/neliq/Documents/NIF/Rothamsted/shapefiles/area_withCT_all.shp')
```
