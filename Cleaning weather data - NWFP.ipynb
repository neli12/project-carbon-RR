```python
# Load libraries
import pandas as pd
import numpy as np
import os
import glob
import csv
```

```python
# Create a list with column names
cols = ['ob_end_time', 'max_air_temp', 'min_air_temp']
```

```python
# Read files
all_files = [i for i in glob.glob('*.csv')]

#combine all files
concat_csv = pd.concat([pd.read_csv(f, skiprows=90, usecols = cols)[:-1] for f in all_files])
concat_csv.rename(columns = {'ob_end_time': 'date'},  inplace = True)
concat_csv
```

```python
concat_csv['date'] = pd.to_datetime(concat_csv['date'])
concat_csv.info()
```

```python
concat_csv['year'] = concat_csv['date'].dt.year
concat_csv['month'] = concat_csv['date'].dt.month
concat_csv['day'] = concat_csv['date'].dt.day
concat_csv['day_ofthe_year'] = concat_csv['date'].apply(lambda x:x.dayofyear)
concat_csv
```

```python
#Reorder columns
cols = ['day', 'month', 'year', 'day_ofthe_year', 'max_air_temp', 'min_air_temp']
concat_csv = concat_csv[cols]
concat_csv
```

```python
#Export to csv
concat_csv.to_csv( "temp1982-2021.csv", sep = "\t", index=False, encoding='utf-8-sig')
```
