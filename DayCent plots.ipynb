### Analysing SPIN UP results from DayCent - North WYke site Burrows field


##### Main outptus
'agcprd: Aboveground C production for the grass/crop for the current month year (g C m-2)'  
'bgcjprd: Juvenile fine root C production for the grass/crop over the last completed growing season (g C m-2 yr-1)',   
'bgcmprd: Mature fine root C production for the grass/crop over the last completed growing season (g C m2 yr)',   
'som1c(1):  total C in active surface organic matter pool (g C m-2)',   
'som1c(2): total C in active soil organic matter pool (g C m-2)',  
'som2c(1): total C in slow surface organic matter pool (g C m-2)' ,  
'som2c(2): total C in slow soil organic matter pool (g C m-2)',  
'som3c: total C in passive soil organic matter pool (g C m-2)',   
'somsc: sum of labeled and unlabeled C from som1c(2), som2c(2), and som3c(g C m2)',   
'somtc:  total soil C including belowground structural and metabolic: somsc + strucc(2) + metabc(2) (g C m-2)',   
'sdrema: annual accumulator of C removed from standing dead during grazing/fire for grass/crop (g C m-2yr-1)',   
'shrema: annual accumulator of C removed from live shoots during grazing/fire for grass/crop (g C m-2yr-1)',  
'stdedc: C in standing dead material for grass/crop (g C m-2)',   
'aglivc: C in aboveground live for grass/crop (g C m-2)',   
'stream(2):  N from mineral leaching of stream flow (base flow + storm flow) (g N m-2)',   
'resp(1):  annual unlabeled CO2 respiration from decomposition (g C m-2)',   
'tminrl(1): total mineral N summed across layers (g N m-2)',   
'minerl(1,1): mineral N content for layer 1 (g N m-2)',  
'minerl(2,1): mineral N content for layer 2 (g N m-2)',   
'minerl(3,1): mineral N content for layer 3 (g N m-2)',   
'minerl(4,1): mineral N content for layer 4 (g N m-2)',   
'minerl(5,1): mineral N content for layer 5 (g N m-2)',  
'minerl(6,1): mineral N content for layer 6 (g N m-2)',   
'minerl(7,1): mineral N content for layer 7 (g N m-2)',   
'minerl(8,1): mineral N content for layer 8 (g N m-2)',   
'minerl(9,1): mineral N content for layer 9 (g N m-2)',  
'minerl(10,1): mineral N content for layer 10 (g N m-2)'  

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read file and show info
soilNW = pd.read_csv("NW6_spin.lis", sep = "\s+")[:-1]
soilNW['time'] = soilNW['time'].astype(int)
soilNW.head()
```

### Aboveground production

```python
# Plot
plt.figure(figsize=(14,5))
plt.subplot(1, 2, 1)
soilNW['agcprd'].plot(color = 'black', linewidth=0.4)
plt.xlim(0, 2000)
#plt.ylim(-0.04, 0.04)
plt.ylabel('Aboveground C production (g C m2)')
plt.title("Aboveground production")
plt.xlabel('Year')
plt.grid(True)

plt.subplot(1, 2, 2)
soilNW['bgcjprd'].plot(color = 'black', linewidth=0.4)
plt.xlim(0, 2000)
#plt.ylim(-0.04, 0.04)
plt.ylabel('Root C production (g C m2 yr)')
plt.title("Root C production")
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

### Aboveground production - Zoom to year 1100 to 1982

```python
# Plot
plt.figure(figsize=(14,5))
plt.subplot(1, 2, 1)
soilNW['agcprd'].plot(color = 'black', linewidth=0.6)
plt.xlim(1100, 2000)
plt.ylabel('Aboveground C production (g C m2)')
plt.title("Aboveground C production from year 1100 to 2000")
plt.xlabel('Year')
plt.grid(True)

plt.subplot(1, 2, 2)
soilNW['bgcjprd'].plot(color = 'black', linewidth=0.6)
plt.xlim(1100, 2000)
plt.ylabel('Root C production (g C m2 yr)')
plt.title("Root C production from year 1100 to 2000")
plt.xlabel('Year')
plt.grid(True)
plt.show()
```
