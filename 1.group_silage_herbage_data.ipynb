```python
# Import libraries
import pandas as pd
import numpy as np
import geopandas as gp
import matplotlib.pyplot as plt
import glob
import folium
```

```python
# List all herbage data
path_herbage = 'C:/Users/neliq/Documents/NIF/Rothamsted/1.2.Herbage_data'
files_herbage = glob.glob(path_herbage + '/' + '*Herbage Survey*.csv')
files_herbage
```

```python
# List all silage files
path_silage = 'C:/Users/neliq/Documents/NIF/Rothamsted/1.1.Silage_data'
files_silage = glob.glob(path_silage + '/' + '*Silage*.csv')
files_silage
```

```python
# Join all silage datasets
files_concat_silage = pd.concat([pd.read_csv(f) for f in files_silage])
files_concat_silage.head()
```

```python
# Join all herbage datasets
files_concat_herbage = pd.concat([pd.read_csv(f) for f in files_herbage])
files_concat_herbage.head()
```

```python
# Check field names
#files_concat_herbage['Field'].unique().tolist()
files_concat_silage['Field'].unique().tolist()
```

```python
# Select only Field 'Burrows'
fields_silage = ['Burrows', 'Burrows, Burrows']
filesBurrows_silage = files_concat_silage[files_concat_silage.Field.isin(fields_silage)].reset_index(drop = True)
filesBurrows_silage.drop(['SP_ID', 'sample_distance (m)', 'Start_Sample_Date', 'Silage Dry Matter Yield Quality Last Modified'], 
                  axis = 1, inplace = True)
filesBurrows_silage.head()
```

```python
# Select only Field 'Burrows'
fields_herbage = ['Burrows']
filesBurrows_herbage = files_concat_herbage[files_concat_herbage.Field.isin(fields_herbage)].reset_index(drop = True)
filesBurrows_herbage.head()
```

```python
# Check dates
filesBurrows_herbage['Sample_date'].unique()
filesBurrows_silage['Sample_date'].unique()
```

```python
# Group by dates and convert to g_m2
burrowsSilage_by_date = filesBurrows_silage.groupby(['Sample_date']).mean().reset_index(drop = False).round(2)
burrowsHerbage_by_date = filesBurrows_herbage.groupby(['Sample_date']).mean().reset_index(drop = False).round(2)

burrowsSilage_by_date['silage_gm2'] = (burrowsSilage_by_date['Silage Dry Matter Yield (kg/ha)'] / 10).round(2)
burrowsSilage_by_date
```

```python
# Export
burrowsSilage_by_date.to_csv("C:/Users/neliq/Documents/NIF/Rothamsted/1.1.Silage_data/burrowsSilage_by_date.csv", 
                             index = False)
burrowsHerbage_by_date.to_csv("C:/Users/neliq/Documents/NIF/Rothamsted/1.2.Herbage_data/burrowsHerbage_by_date.csv", 
                              index = False)
```
